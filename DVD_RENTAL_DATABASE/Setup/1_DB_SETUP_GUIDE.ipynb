<p><strong>Option 1: Load the sample database using the psql</strong></p>

<p>To create a database and load data from an archive file, follow these steps:</p>

<ol>
  <li>Connect to the PostgreSQL database server using <code>psql</code> or pgAdmin.</li>
  <li>Create a blank database called <code>dvdrental</code>.</li>
  <li>Load data from the sample database file into the <code>dvdrental</code> database using <code>pg_restore</code>.</li>
</ol>

<h3>1) Create the dvdrental database</h3>
<p>Open Command Prompt (Windows) or Terminal (Linux/Mac) and connect:</p>
<pre><code>psql -U postgres
</code></pre>

<p>Enter the password for the <code>postgres</code> user. After successful login, the prompt looks like:</p>
<pre><code>postgres=#
</code></pre>

<p>Create the <code>dvdrental</code> database:</p>
<pre><code>CREATE DATABASE dvdrental;
</code></pre>

<p>Output:</p>
<pre><code>CREATE DATABASE
</code></pre>

<h3>2) Download and load the sample database</h3>
<p>Download <code>dvdrental.zip</code> and extract <code>dvdrental.tar</code> to a directory, for example:</p>
<pre><code>D:\sampledb\postgres\dvdrental.tar
</code></pre>

<p>Load the database using <code>pg_restore</code> in the command line:</p>
<pre><code>pg_restore -U postgres -d dvdrental D:\sampledb\postgres\dvdrental.tar
</code></pre>

<p><strong>Explanation:</strong></p>
<ul>
  <li><code>-U postgres</code> ‚Üí Connect as postgres user</li>
  <li><code>-d dvdrental</code> ‚Üí Target database name</li>
</ul>

<p>Enter the postgres password when prompted. Data will be loaded into <code>dvdrental</code>.</p>

<h3>3) Verify the sample database</h3>
<p>Connect to PostgreSQL server and switch to the dvdrental database:</p>
<pre><code>psql -U postgres
\c dvdrental
</code></pre>

<p>The prompt changes to:</p>
<pre><code>dvdrental=#
</code></pre>

<p>List all tables:</p>
<pre><code>\dt
</code></pre>

    <p><strong>Option 2: Restore in pgAdmin GUI (No Command Needed)</strong></p>

<ol> <li><strong>Open pgAdmin</strong><br> Launch from Start Menu ‚Üí pgAdmin 4 </li> <li><strong>Connect to Server</strong><br> Enter your master password </li> <li><strong>Navigate to the Database Section</strong><br> Expand: Servers &gt; PostgreSQL </li> <li><strong>Right-click on "Databases"</strong><br> Click <em>Create</em> ‚Üí <em>Database...</em> </li> <li><strong>Fill in the Form:</strong> <ul> <li><strong>Database name:</strong>  <em>dvdrental</em></li> <li><strong>Owner:</strong> Leave as <em>postgres</em> or choose another user</li> <li>Leave other settings default unless you need advanced configuration</li> </ul> </li> <li><strong>Click Save</strong></li> </ol> <p>Your new database will now appear in the left panel under <strong>Databases</strong>.</p>
<p><strong>NOW:</strong></p>
<ol>
  <li>Open pgAdmin</li>
  <li>Right-click on the database ‚Üí <strong>Restore</strong></li>
  <li>Select <strong>Format = Custom or Tar</strong></li>
  <li>Browse to your <code>.tar</code> file</li>
  <li>Click <strong>Restore</strong></li>
